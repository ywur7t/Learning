```python
import redis
import json
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd
```

```python
r = redis.Redis(
    host="localhost",
    port=6379,
    db=0,
    decode_responses=True,
    password='example'
)
```

```python
df = pq.read_table("./data/prepared_data.parquet").to_pandas()
```

```python
for i, row in df.iterrows():
    key = f"trade:{i}"
    value = row.to_dict()
    
    r.set(key, json.dumps(value))
```

```python
print(r.get("trade:0"))
```
